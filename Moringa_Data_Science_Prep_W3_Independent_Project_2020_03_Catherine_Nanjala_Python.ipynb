```
#import pandas and numpy library
import pandas as pd
import numpy as np
```

```
#loading all the datasets while specifying the column names
#loading the first telcom dataset
df0 = pd.read_csv('Telcom_dataset.csv')
df0
df0.columns = ['PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE','DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
df0
```

```
#load the second telecom dataset
df1 = pd.read_csv('Telcom_dataset2.csv',delimiter=',')
df1
df1.columns = ['PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE','DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
df1
```

```
#loading the third telecom dataset
df2 = pd.read_csv('Telcom_dataset3.csv',delimiter=',')
df2
df2.columns = ['PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE','DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
df2
```

```
#loading the cells_geo dataset
df_geo = pd.read_csv('cells_geo.csv', delimiter=';')
df_geo
```

```
#merging all the datasets
df = pd.concat((df0, df2, df1, df_geo), axis=0, ignore_index=True, sort=False)
df
```

```
#Dropping the columns that are not needed
df_final = df.drop(['COUNTRY_A','COUNTRY_B','CELL_ON_SITE'], axis=1, inplace = False)
df_final
```

```
#converting the DATE_TIME column to datetime data type
df_final['DATE_TIME'] = pd.to_datetime(df_final['DATE_TIME'])
```

```
#Finding out information on the now merged dataset
df_final.info()
```
